1.  What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

- Type I error (a) occurs when we  **reject the null hypothesis when it is actually true**- False Positive.
 Example: Declaring a medicine effective when it actually does nothing.

- Type II error (b) occurs when  we **fail to reject the null hypothesis when it is actually false**- Flase negative.
 Example: Declaring a medicine ineffective when it actually works.

**Impact on decision-making**:

>>- Type I error- Unnecessary costs, wrong actions, loss of trust (e.g., convicting an  innocent person).

>>- Type II error- Missed opportunities, continued harm (e.g., releasing a guilty person or not detecting a disease).  Researches balance both using significance level (a, usually 0.05) and power of the test (1-b).


2. :What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

- The **p-value** is the probability of obtaining test results at least as extreme as the observed results, **assuming that he nuall hypothesis(**Ho**) is true**.

**Interpretation**:
>>- Small p-value( ‚â§ a, typically  ‚â§ 0.05)- Strong evidence **against** Ho - Reject null hypothesis.
>>- Large p-value(> 0.05) - Weak evidence against Ho - Fail to reject null hypothesis.

**Important points**:
>>- p-value is **Not** the probability that Ho is true,
>>- It is **Not** the probability of making a Type I error.
>>- Lower p-value does **not** mean the  effect is practically significant (only statistically significent).

Example:

p=0.03 at a =0.05 - Reject Ho (result is statistically significent).

p=0.38 - Do not reject Ho(not enough evidence).

4. What is a confidence interval, and how does the margin of error influence its width and interpretation?

- A **confidence interval (CI)** is a range of values that is likely to contain the true population parameter with a certain level of confidence (usually 95%).

**Formula:**
         CI = Point estimate ¬± margin of error

Ex- 95% CI: 45% ¬± 3% means we are 95% confident that the true population proportion lies between 42% and 48%.

The Margin of Error(MOE): is the ¬± value in the confidence intervel. it measures the maximum expected difference between the sample estimate and the true population value.

**How margin of error affects the width and interpretation:**

1. **Widthof the interval**
-> The confidence interval width = 2 x margin of error

-> Larger MOE- Wider interval - less precise estimate.

-> Smaller MOE- narrower interval - more precise estimate.

2. **Factors that change the margin of error (and thus width):**
- Confidence level: Higher confidence (99%  vs 95%) - larger MOE- wider interval.
- Sample size(n): Larger sample- smaller MOE - narrower intervel
- Population variability( p or a) : More variability - larger MOE- wider interval.

3.  Interpretation trade-off
-> A Wider interval(large MOE) gives higher confidence but less precision

-> A Narrower interval(small MOE) gives more precision but lower confidence.

The margin of error directly determiner the width of the confidence interval -bigger margin means wider interval and less precise estimate, while smaller margin means narrower interval and more precise estimate. at the chosen confidence level.

5.  Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

- ANOVA (Analysis of Variance) is a statistical method used to test whether there are significant differences between the means of three or more independent groups simultaneously. Its main purpose is to determine if at least one group mean is statistically different from the others, while controlling the overall type I error rate.
**Key Assumptions of One-Way ANOVA:**

>>- Independence: Observations within and between groups are independent.
>>- Normality: Each group's data should be approximately normally distributed.
>>- Homogeneity of Variance(Homosecedasticity): All groups have equal population variances (can be tested using Levene's or Barlett's test).

**How  ANOVA extends hypothesis  beyond two groups:**

6. Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.



```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data (average height of 30 students in cm)
date = np.array([165, 168, 170, 172, 167, 169, 171, 166, 173, 175, 164, 162, 174, 176, 163, 168])

# Known population parameters
pop_mean = 170 # H0: population mean = 170 cm
pop_std = 10 # population standard deviation
alpha = 0.05 # significance level

# Calculations
sample_mean = np.mean(date)
n = len(date)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_stat)))

# Critical value
critical_value = stats.norm.ppf(1 - alpha/2)

print(f"Sample Mean: {sample_mean: 3f} cm")
print(f"Z-statistic: {z_stat: 4f}")
print(f"P-value: {p_value: 4f}")
print(f"Critical Value: {critical_value: 3f}")

if abs(z_stat) > critical_value:
  print("Reject the null hypothesis")
else:
  print("Fail to reject the null hypothesis")
```

7. Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Simulate 1000 trials of binomial(n=10, p=0.5)
data = np.random.binomial(n=10, p=0.5, size=1000)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=range(0,12), alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)\n1000 simulation', fontsize=14)
plt.xticks(range(0,11))
plt.grid(axis='y', alpha=0.3)

# Add theoretical probabilities as text
for i in range(11):
  prob = stats.binom.pmf(i, n=10, p=0.5)
  plt.text(i, plt.ylim()[1]*0.9, f'P({i})={prob:.3f}', ha='center', va='bottom',fontsize=9, rotation=98)
plt.show()

# Print basic statistics
print(f"Mean: {data.mean(): 3f} (theoretical = {10*0.5})")
print(f"Variance: {data.var(): 3f} (theoretical = {10*0.5*0.5})")
```

8. Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Step 1: Choose a clearly NON-NORMAL distribution - Exponential (heavily skewed)
population = np.random.exponential(scale=1, size=10000)

# Plot the population (highly skewed)
plt.figure(figsize=(15, 10)) # Increased figure size for 2x3 grid

plt.subplot(2, 3, 1) # Changed to 2x3 grid, subplot 1
plt.hist(population, bins=50, color='red', alpha=0.7, edgecolor='black')
plt.title('Population: Exponential Distribution (Skewed)', fontsize=12)
plt.xlabel('Values')
plt.ylabel('Frequency')

# Step 2: Demonstrate CLT by taking sample means with increasing sample size
sample_sizes = [1, 5, 10, 30, 100]
colors = ['blue', 'green', 'orange', 'purple', 'brown']
num_samples = 5000 # Increased number of samples

for i, n in enumerate(sample_sizes):
  sample_means = []
  for _ in range(num_samples): # Use num_samples
    sample = np.random.choice(population, size=n)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

  plt.subplot(2, 3, i+2) # Changed to 2x3 grid
  plt.hist(sample_means, bins=50, color=colors[i], alpha=0.8, edgecolor='black')
  plt.title(f'Sample Means (n = {n})\n{num_samples} samples', fontsize=11) # Updated title
  plt.xlabel('Sample Means')
  plt.ylabel('Frequency')
  plt.xlim(0, 6)


plt.suptitle('Central Limit Theorem in Action\nEven from Skewed Exponential - Means become Normal as n Increases', fontsize=16, fontweight='bold') # Changed subtitle to suptitle for overall title
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()

print("CLT DEMONSTRATED: As sample size increases (n=1 - 100), the distribution of sample means approaches a NORMAL distribution, regardless of the original population being highly skewed.")
```

9.  Write a Python function to calculate and visualize the confidence interval
for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval(data, confidence=0.95):
  """
  Calculate and visualize the confidence interval for a sample mean.

  Parameters:
  - data (array-like): The sample data.
  - confidence (float): Confidence level for the interval (default is 0.95).
  """

  data = np.array(data)
  n = len(data)
  mean = np.mean(data)
  std_error = stats.sem(data) # Standard error of the mean

  t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
  margin_of_error = t_value * std_error

  lower_bound = mean - margin_of_error
  upper_bound = mean + margin_of_error

  ci_lower = lower_bound
  ci_upper = upper_bound

  print(f"Sample Mean: {mean: .3f}")
  print(f"Margin of Error: {margin_of_error: .3f}")
  print(f"{int(confidence*100)}% Confidence Interval: [{ci_lower: .3f}, {ci_upper: .3f}]")

  plt.figure(figsize=(8, 4))
  plt.hist(data, bins=20, color='skyblue', alpha=0.7, edgecolor='black')
  plt.axvline(mean, color='red', linestyle='--', label=f'Sample Mean = {mean: .2f}')
  plt.axvline(ci_lower, color='gray', linestyle='--', label=f"Lower CI = {ci_lower: .2f}")
  plt.axvline(ci_upper, color='gray', linestyle='--', label=f"Upper CI = {ci_upper: .2f}")
  plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
  plt.xlabel('Sample Values')
  plt.ylabel("Frequency")
  plt.legend()
  plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
confidence_interval(sample_data, confidence=0.95)
```

10. Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy import stats

# Step 1: Observed frequencies (e.g., 120 students' favorite subject)
observed = np.array([35, 25, 20, 30, 10]) # Math, Science, English, History, Arts

# Step 2: Expected frequencies (if uniform preference)
expected = np.array([24, 24, 24, 24, 24]) # 120 students / 5 subject =24 each

# Step 3: Perform Chi-square Goodness-of test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Step 4: Results
print(f"CHI-SQUARE GOODNESS-OF-FIT TEST")
print("H0: The obsereved distribution fits the expected uniform distribution")
print("H1: The distribution does not NOT fit")
print(f"Observed: {observed}")
print(f"Expected: {expected}")
print(f"Chi-square Statistic: {chi2_stat: 3f}")
print(f"P-value: {p_value: 4f}")

alpha = 0.05
if p_value < alpha:
  print("DECISION: Reject H0")
  print("Conclusion: Students do NOT have uniform preference across subject.")
  print("  - Math (35) and History (30) are more popular than Arts (10) !")
else:
  print("DECISION: Fail to reject H0")
  print("Conclusion: No evidence against uniform preference.")

# Optional : Visualize
import matplotlib.pyplot as plt
subject = ['Math', 'Science', 'English', 'History', 'Arts']
x = np.arange(len(subject))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, observed, width, label='Observed', color='skyblue', edgecolor='black')
plt.bar(x + width/2, expected, width, label='Expected', color='salmon', alpha=0.7)
plt.xticks(x, subject)
plt.title('Observed vs Expected Subject Preferences\nChi-square Test Result', fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```
